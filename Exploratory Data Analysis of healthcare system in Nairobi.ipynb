```python
#droping the null columns from dataframe
df = df.drop(columns=['ANC','BEOC','BLOOD','CAES SEC','CEOC','EPI','GROWM','OPD', 'OUTREACH', 'PMTCT','RAD/XRAY','RHTC/RHDC','TB DIAG',
       'TB LABS','TB TREAT','YOUTH','HCT'])
```

```python
#filled null columns with avg beds and cots
df['Beds'].fillna(df['Beds'].mean(),inplace=True)
df['Cots'].fillna(df['Cots'].mean(),inplace=True)
```

```python
#converting to appropriate datatypes
df[['Beds','Cots']]=df[['Beds','Cots']].astype('int')
```

```python
df.info()
```

```python
#statastical summery of data
df.describe()
```

```python
# Ensured that categorical variables like 'Owner' and 'Sub-county' are encoded properly.
df['Owner'] = df['Owner'].astype('category')
df['Sub Location'] = df['Sub Location'].astype('category')
```

```python
#converted operational status as boolean
df['Operational Status'] = df['Operational Status'].apply(lambda x: True if x == 'Operational' else False)
df
```

```python
# Task 2.2: Visualize the distribution of healthcare facilities by type and ownership
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Type', hue='Owner')
plt.title("Distribution of Healthcare Facilities according to Type and Ownership")
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert boolean values in 'Operational Status' to 'Operational' and 'Non-Operational'
df['Operational Status'] = df['Operational Status'].map({True:'Operational', False:'Non-Operational'})
```

```python
# Task 2.3: Compare operational vs. non-operational facilities across different regions
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='County', hue='Operational Status')
plt.title("Operational vs Non-Operational Facilities across Counties")
plt.xticks(rotation=90)
plt.show()
```

```python
non_operational = df[df['Operational Status'] == 'Non-Operational']
non_operational_count = non_operational['County'].value_counts()
print("Sub-counties with highest non-operational facilities:")
print(non_operational_count.head()) 
```

```python
# Task 3.2: Analyze which sectors own the most operational facilities
operational = df[df['Operational Status'] == 'Operational']
ownership_count = operational['Owner'].value_counts()
print("Sectors owning the most operational facilities:")
print(ownership_count)
```

```python
# Task 4.1: Bar chart showing facility ownership distribution by sub-county
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='County', hue='Owner')
plt.title("Facility Ownership Distribution by County")
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 4.3: Pie chart showing the proportion of different healthcare facility types
facility_types = df['Type'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(facility_types, labels=facility_types.index, autopct='%2.1f%%', startangle=140)
plt.title("Distribution according to Healthcare Facility Types")
plt.show()
```
